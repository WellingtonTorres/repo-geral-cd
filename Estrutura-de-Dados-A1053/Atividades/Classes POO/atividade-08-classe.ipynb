# **Atividade para entrega**

Desenvolver um script em Python aplicando os pilares de **Orientação a Objetos** (encapsulamento, herança e polimorfismo):

1. Criar uma classe chamada `Funcionario` com os atributos privados `__nome` e `__salario_base`, e métodos *getter* para cada um.
2. Criar um método `calcular_pagamento()` que retorne o salário base.
3. Criar uma classe filha `Professor` que herde de `Funcionario` e possua o atributo adicional `horas_extras_orientacao`.
4. Aplicar **Polimorfismo**: sobrescrever o método `calcular_pagamento()` na classe `Professor` para retornar o salário base somado a (`horas_extras_orientacao` * 50 reais).

---

```python
## 1º Crie uma classe chamada Funcionario que tenha os atributos privados __nome e __salario_base. Crie métodos getter para eles.
class Funcionario:
    def __init__(self, nome, salario_base):
        self.__nome = nome
        self.__salario_base = salario_base

    def get_nome(self):
        return self.__nome

    def get_salario_base(self):
        return self.__salario_base

    ## 2º Crie um método chamado calcular_pagamento() que simplesmente retorne o salário base.
    def calcular_pagamento(self):
        return self.__salario_base


## 3º Crie uma classe filha chamada Professor que herde de Funcionario . O Professor possui um atributo adicional chamado horas_extras_orientacao.
class Professor(Funcionario):
    def __init__(self, nome, salario_base, horas_extras_orientacao):
        
        super().__init__(nome, salario_base)
        self.horas_extras_orientacao = horas_extras_orientacao

    ## 4º Aplique o Polimorfismo: sobrescreva o método calcular_pagamento() na classe Professor para que ele retorne o salário base somado a (horas_extras * 50 reais).
    def calcular_pagamento(self):
        return self.get_salario_base() + (self.horas_extras_orientacao * 50)


## Demostração
func = Funcionario('Ana Silva', 3000)
prof = Professor('Carlos Souza', 4000, 10)

print(f'Funcionário: {func.get_nome()}')
print(f'Pagamento: R$ {func.calcular_pagamento():.2f}\n')

print(f'Professor: {prof.get_nome()}')
print(f'Horas extras de orientação: {prof.horas_extras_orientacao}')
print(f'Pagamento: R$ {prof.calcular_pagamento():.2f}')
```
