### **1º Exercício**

1. Escreva um programa em Python que:
a)
Leia do usuário uma sequência de números inteiros separados por espaço e armazene-os em uma lista.
b)
Realize as seguintes operações:
i.
Exiba a lista original.
ii.
Crie uma nova lista apenas com os números pares, mantendo a ordem original.
iii.
Crie outra lista com os elementos da lista original sem o primeiro e o último elemento (use fatiamento).
iv.
Calcule a média dos valores da lista original e exiba.
v.
Encontre o segundo maior valor da lista original.
c)
Mostre todas as listas e os resultados calculados.

```python
sequencia = input("Digite uma sequência de números inteiros separados por espaço: ")
dadosLista = []
listaPares = []
listaFatiada = []
mediaLista = 0
maiorValor = 0
segMaiorValor = 0
dadosTratados = []

for atual in sequencia.split():
    numero = int(atual)
    dadosLista.append(numero)
    if numero % 2 == 0:
        listaPares.append(numero)
        
listaFatiada = dadosLista[1:len(dadosLista)-1]
mediaLista = sum(dadosLista) / len(dadosLista)

# Segundo maior valor
dadosTratados = list(set(dadosLista)) #remove duplucados
#print(f'Dados tratados: {dadosTratados}')
maiorValor = max(dadosTratados)
dadosTratados.remove(maiorValor)
segMaiorValor = max(dadosTratados)
        
print(f'Lista original: {dadosLista}')
print(f'Lista de pares: {listaPares}')
print(f'Lista fatiada: {listaFatiada}')
print(f'Média valores Lista original: {mediaLista:.2f}')
print(f'Maior valor: {maiorValor}')
print(f'Segundo maior valor: {segMaiorValor}')
```

### **5º Exercício**

5. Implemente uma função que recebe um número inteiro N e retorna uma lista com todos 
os números primos de 1 até N.

```python
num1 = 10
def listaPrimos(num :int):
    primos = []
    for n in range(2, num + 1):
        primo = True
        for i in range(2, n):
            if n % i == 0:
                primo = False
                break        
        if primo:
            primos.append(n)
    
    return primos
print(f'o numero {num1} gerou a lista de primos: {listaPrimos(num1)}')
```

### **6º Exercício**

6. Crie uma função que recebe um dicionário com nomes como chave e idades como valor, 
e retorne o nome da pessoa mais velha.

```python
#solução mais simples
pessoas = {'Ana': 18,
           'Carlos': 38,
           'Eduarda': 29,
           'João': 89}

def maisVelho(pessoas: dict):
    maior = 0
    nome = ''
    for chave, valor in pessoas.items():
        if valor > maior:
            maior = valor
            nome = chave
    print(f'A pessoa mais velha é {nome} com {valor} ano(s)')

maisVelho(pessoas)

#solucao IA
def mais_velho(pessoas: dict):
    nome = max(pessoas, key=pessoas.get)
    return nome, pessoas[nome]

nome, idade = mais_velho(pessoas)

print(f'A pessoa mais velha é {nome} com {idade} ano(s)')
```

### **9º Exercício**

9. Desenvolva uma função que aceite uma string e retorne  True se ela for um palíndromo 
(lê-se igual de frente para trás e vice-versa), e False caso contrário

```python
palavra = 'Socorram me subi no onibus em Marrocos'.lower()
def validaPalindromo(texto :str):
    texto = texto.replace(" ", "")
    textoinvertido = texto[::-1]
    return texto == textoinvertido

print(f'{validaPalindromo(palavra)}')
```

### **10º Exercício**

10. Implemente  uma  função  que  recebe  uma  lista  de  números  e  retorne  o  segundo  maior 
número da lista, considerando listas com pelo menos dois elementos

```python
numeros = [2,5,3,1,10,12]

def segundoMaiorA(lista: list):
    maior = max(lista)
    while maior in lista:
        lista.remove(maior)
    return max(lista)

print(f'{segundoMaiorA(numeros)}')
```
