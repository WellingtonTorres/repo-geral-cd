# Programação Orientada a Objetos (POO) em Python - Básico

O paradigma **Programação Orientada a Objetos (POO)** é fundamental para organizar códigos complexos, sejam eles scripts de análise para Ciência de Dados, integrações de Inteligência Artificial ou qualquer outro sistema que requer software.

## 1. Introdução à POO

A Programação Orientada a Objetos é um paradigma focado em modelar o mundo real dentro do código. Em vez de pensarmos apenas em "variáveis" e "funções" soltas (como no paradigma estruturado), pensamos em **entidades** que possuem características e comportamentos.


### 1.1 Objetos, Classes, Atributos e Métodos

* **Classe:** É o "molde" ou a "planta baixa". Define as características e os comportamentos que algo terá.
* **Objeto:** É a instância, ou seja, o "algo" criado a partir do molde.
* **Atributos:** São as características do objeto (variáveis internas).
* **Métodos:** São as ações que o objeto pode realizar (funções internas).

Vamos criar nossa primeira classe. Imagine que estamos programando um braço robótico ou um robô autônomo.

```python
# definindo a Classe
class Robo:
    # O método __init__ é o construtor, chamado automaticamente ao criar o objeto
    def __init__(self, nome, tipo):
        self.nome = nome       # atributo
        self.tipo = tipo       # atributo
        self.ligado = False    # atributo

    # método (comportamento ou ação)
    def ligar(self):
        self.ligado = True
        print(f'[{self.nome}] Sistema iniciado')

    # método (comportamento ou ação)
    def desligar(self):
        self.ligado = False
        print(f'[{self.nome}] Sistema desligado')

# Criando objetos (ou "instâncias")
meu_robo = Robo(nome='R2-D2', tipo='Astromech')
robo_industrial = Robo(nome='Braço XYZ', tipo='Manufatura')

# usando o objeto
meu_robo.ligar()
print(f'O robô {meu_robo.nome} está ligado? {meu_robo.ligado}')
```

## Entendendo o `self`

Em POO com Python, o `self` é uma referência à **própria instância** (o objeto) que está sendo manipulada naquele momento. De forma simples: é a maneira de um objeto dizer **"eu"** ou **"meu"**.

Sempre que você cria uma classe, você está criando um "molde". Quando você cria um objeto a partir desse molde, você tem uma "instância". O `self` serve para que o Python saiba de qual instância específica ele deve ler ou alterar os dados.

## 2. Outro Exemplo - A Classe Matriz
Na Aula 05 de Estrutura de Dados, começamos a ver como organizar dados tabulares de forma eficiente. Em vez de lidarmos com listas de listas soltas no código, podemos "empacotar" a lógica matemática e algorítmica dentro de uma classe `Matriz`.

Isso reduz a carga cognitiva de quem for usar nossa estrutura depois.

```python
class Matriz:
    def __init__(self, linhas, colunas):
        self.linhas = linhas
        self.colunas = colunas
        # compreensão de listas em Python para criar a matriz inicializada com zeros
        self.dados = [[0 for _ in range(colunas)] for _ in range(linhas)]

    def preencher_valor(self, i, j, valor):
        #Define um valor específico na linha i e coluna j
        if 0 <= i < self.linhas and 0 <= j < self.colunas:
            self.dados[i][j] = valor
        else:
            print('Erro: Índice fora dos limites da matriz.')

    def mostrar(self):
        #exibe a matriz formatada no console
        for linha in self.dados:
            print(linha)


m1 = Matriz(3, 3)
m1.preencher_valor(0, 0, 10)
m1.preencher_valor(1, 1, 20)
m1.preencher_valor(2, 2, 30)

print('Matriz Original:')
m1.mostrar()
```

## 3. Encapsulamento


O encapsulamento é o princípio de "esconder" os detalhes internos de funcionamento de uma classe. Se deixarmos o atributo `dados` exposto, um usuário desavisado pode apagar a matriz sem querer (ex: `m1.dados = None`).

Em Python, indicamos que um atributo é **privado** colocando dois underlines (`__`) antes do nome. O acesso a esses atributos passa a ser feito através de métodos específicos, chamados *Getters* (para pegar) e *Setters* (para alterar).

```python
class MatrizEncapsulada:
    def __init__(self, linhas, colunas):
        # os atributos agora são privados
        self.__linhas = linhas
        self.__colunas = colunas
        self.__dados = [[0 for _ in range(colunas)] for _ in range(linhas)]

    # Getter para acessar a matriz de fora de forma segura
    def get_dados(self):
        return self.__dados

    def get_dimensoes(self):
        return self.__linhas, self.__colunas

    def preencher_valor(self, i, j, valor):
        if 0 <= i < self.__linhas and 0 <= j < self.__colunas:
            self.__dados[i][j] = valor

    def mostrar(self):
        for linha in self.__dados:
            print(linha)

m2 = MatrizEncapsulada(2, 2)
m2.preencher_valor(0, 0, 5)


print('\nMatriz Encapsulada (acesso seguro via getter):', m2.get_dados())
```

## 4. Herança e Polimorfismo

Quando modelamos sistemas complexos, frequentemente notamos que algumas classes compartilham a mesma base, mas precisam de algumas características extras. É aqui que entram a Herança e o Polimorfismo.


* **Herança:** Permite criar uma "Classe Filha" que herda atributos e métodos de uma "Classe Mãe". Reaproveitamento de código!
* **Polimorfismo:** Permite que métodos com o mesmo nome se comportem de forma diferente em classes diferentes.

```python
# Herança: MatrizQuadrada herda tudo da nossa classe Matriz original
class MatrizQuadrada(Matriz):
    def __init__(self, ordem):
        # chama o construtor da classe Mãe (Matriz), passando a ordem para linhas e colunas
        super().__init__(ordem, ordem)
        self.ordem = ordem

    # Polimorfismo: sobrescrevemos o método 'mostrar' para indicar que é Quadrada
    def mostrar(self):
        print(f'--- Exibindo Matriz Quadrada de Ordem {self.ordem} ---')
        super().mostrar() # Usa a lógica de exibição da classe mãe

    # Método exclusivo da classe filha
    def preencher_diagonal_principal(self, valor):
        for i in range(self.ordem):
            self.preencher_valor(i, i, valor)

print('Testando Herança e Polimorfismo:')
mq = MatrizQuadrada(4)
mq.preencher_diagonal_principal(1)
mq.mostrar()
```

## Entendendo a função `super()`

Quando trabalhamos com **Herança** na Programação Orientada a Objetos, criamos uma Classe Filha que herda características de uma Classe Mãe (também chamada de *Superclasse*).

No entanto, muitas vezes a Classe Filha precisa ter o seu próprio método construtor (`__init__`) para receber atributos novos. Quando fazemos isso, o Python "sobrescreve" o construtor da mãe. Para não perdermos a lógica que já estava pronta na classe Mãe, usamos a função **`super()`**.

Basicamente, o `super()` é um atalho que diz ao Python: **"Vá até a minha classe Mãe e execute este método lá"**.

### Analogia da receita de bolo
Imagine que a Classe Mãe é uma receita de **Bolo Básico** (misturar farinha, ovos e leite, e assar).
A Classe Filha é uma receita de **Bolo de Chocolate**.

Em vez de você reescrever todos os passos de misturar farinha e ovos na receita do bolo de chocolate, você simplesmente diz: *"Faça a receita do Bolo Básico (`super()`) e, no final, adicione chocolate em pó"*.
